```python
import pandas as pd
import googlemaps
from datetime import datetime

# Group API
api = 'AIzaSyBYudclyrbOoorI1r8zepAb1vS-Z6Pqpng'
# Set up API
gmaps = googlemaps.Client(key= api)

# Create new columns "latitud" and "longitud"
df['Latitud'] = None
df['Longitud'] = None

#Creamos una nueva columna llamada "Dirección completa" y Concatenamos las direcciones completas con la finalidad de 
#capturar zonas con mismo nombre en distintos países/zonas
df['Direccion Completa'] = df['Direccion'] + ', ' + df['DISTRITO'] + ', ' + df['PROVINCIA'] + ', ' + df['DEPARTAMENTO']

# Iterando sobre las filas del DataFrame
for index, row in df.iterrows():
    # Obtiene la dirección de la fila
    # Concatanete directions with specific sub-levels to increase accuraccy
    nuevo_df = pd.DataFrame({
        'Direccion Completa': df['Direccion'] + ', ' + df['DISTRITO'] + ', ' + df['PROVINCIA'] + ', ' + df['DEPARTAMENTO'] })

# Iterative process
for index, row in nuevo_df.iterrows():
    # Get direction row
    direccion = row['Direccion Completa']
    
    # Get API direction
    geocode_result = gmaps.geocode(direccion)

    # Check and update the data with the lat and long
    if geocode_result:
        location = geocode_result[0]['geometry']['location']
        df.at[index, 'Latitud'] = location['lat']
        df.at[index, 'Longitud'] = location['lng']

df
```

3. Use Google API to find the driving time (best guess) from all the group members' address and all the LIMA BBVA offices.


```python
# Lista de direcciones de los miembros del grupo (reemplaza con las direcciones reales)
direcciones_miembros = ['Av. Universitaria, Lima 15302', 'C. Chinchero 175, Lima 15024', 'San Jacinto 205, Lima 15023', 'Jr. Huánuco 317, Lima 15003', 'C. Los Tulipanes 432, Lima 15046']

# Extrae las direcciones de las oficinas del BBVA del DataFrame
direcciones_bbva = df['Direccion Completa'].tolist()
```

```python
# Iterar sobre cada combinación de dirección de miembro y oficina del BBVA
for direccion_miembro in direcciones_miembros:
    for direccion_bbva in direcciones_bbva:
        # Realizar la solicitud a la API de Google Maps Directions
        directions_result = gmaps.directions(direccion_miembro,
                                             direccion_bbva,
                                             mode="driving",
                                             departure_time=datetime.now())
        # Extraer y mostrar el tiempo de conducción
        if directions_result:
            duracion = directions_result[0]['legs'][0]['duration']['text']
            print(f"Tiempo de conducción desde {direccion_miembro} hasta {direccion_bbva}: {duracion}")
        else:
            print(f"No se pudo obtener el tiempo de conducción desde {direccion_miembro} hasta {direccion_bbva}")
```

4. Finally, you have to give a report which offices are the closest and furthest to every group member's address.

```python
import re 
distances_info = {}

for direccion_miembro in direcciones_miembros:
    # We create empty variables here so that they reset every time we change to the address of other members
    max_distance = 0
    min_distance = 60
    furthest_bbva = None
    closest_bbva = None

    for direccion_bbva in direcciones_bbva:
        # Realizar la solicitud a la API de Google Maps Directions
        directions_result = gmaps.directions(direccion_miembro,
                                             direccion_bbva,
                                             mode="driving",
                                             departure_time=datetime.now())
        # Extraer y mostrar el tiempo de conducción
        if directions_result:
            duracion = directions_result[0]['legs'][0]['duration']['text']
            
            # Convert distances to integer
            duracion_numeric = int(re.search(r'\d+', duracion).group())

            # Compare and update furthest and closest distances
            if duracion_numeric > max_distance:
                max_distance = duracion_numeric
                furthest_bbva = direccion_bbva

            if duracion_numeric < min_distance:
                min_distance = duracion_numeric
                closest_bbva = direccion_bbva

    # Store information about both furthest and closest offices for the current member
    distances_info[direccion_miembro] = {
        'furthest_distance': max_distance,
        'furthest_bbva': furthest_bbva,
        'closest_distance': min_distance,
        'closest_bbva': closest_bbva
    }


distances_info
```
